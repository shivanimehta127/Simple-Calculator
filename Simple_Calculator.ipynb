```python
def addition(num1, num2):
  return num1 + num2

def substraction(num1, num2):
  if num1 > num2:
    return num1 - num2
  else:
    return num2 - num1

def multiplication(num1, num2):
  return num1 * num2

def division(dividend, divisor):
    return dividend / divisor

def modulus(dividend, divisor):
    return dividend % divisor

def power(base, raise_to_power):
  return base * raise_to_power

def percentage(value, total):
   return (value/total) * 100

def factorial(number):
  if number == 0:
    return 1
  else:
    fact = 1
    temp = number
    while temp > 0:
      fact = fact * temp
      temp = temp - 1
    return fact


def input_calculator(function):
  if function == "addition":
    num1 = int(input("Please enter 1st value "))
    num2 = int(input("Please enter 2nd value "))
    return addition(num1, num2)

  elif function == "substraction":
    num1 = int(input("Please enter 1st value "))
    num2 = int(input("Please enter 2nd value "))
    return substraction(num1, num2)

  elif function == "multiplication":
    num1 = int(input("Please enter 1st value "))
    num2 = int(input("Please enter 2nd value "))
    return multiplication(num1, num2)

  elif function == "division":
    divident = int(input("Please enter divident "))
    divisor = int(input("Please enter divisor "))
    if divisor == 0:
      print("Please enter the input again")
      divident = int(input("Please enter divident "))
      divisor = int(input("Please enter divisor "))
    return division(divident, divisor)

  elif function == "modulus":
    divident = int(input("Please enter divident "))
    divisor = int(input("Please enter divisor "))
    if divisor == 0:
     print("Please enter the input again")
     divident = int(input("Please enter divident "))
     divisor = int(input("Please enter divisor "))
    return modulus(divident, divisor)

  elif function == "percentage":
    value = int(input("Please enter the value whose percentage is to be found "))
    total = int(input("Please enter the total value "))
    if total == 0:
     print("Please enter the input again")
     value = int(input("Please enter the value whose percentage is to be found "))
     total = int(input("Please enter the total value "))
    return percentage(value, total)

  elif function == "power":
    base = int(input("Please enter base "))
    raise_to_power = int(input("Please enter power "))
    return power(base, raise_to_power)

  elif function == "factorial":
    number = int(input("Please enter the number "))
    if number < 0:
      print("Please enter the input again")
      number = int(input("Please enter the number "))
    return factorial(number)

  else:
    print("Please enter valid values")
    calculator()


def calculator():
  operations = {
      1 : "addition",
      2 : "substraction",
      3 : "multiplication",
      4 : "division",
      5 : "modulus",
      6 : "percentage",
      7 : "power",
      8 : "factorial"
  }

  print("Enter the key of operatin you want to perform")
  func = int(input("1 : Addition \n 2 : Substraction \n 3 : Multiplication \n 4 : Division \n 5 : Modulus \n 6 : Percentage \n 7 : Power \n 8 : Factorial \n"))
  if not operations.get(func):
    print("Enter a valid value")
    calculator()
  value = input_calculator(operations.get(func))
  return value
```
